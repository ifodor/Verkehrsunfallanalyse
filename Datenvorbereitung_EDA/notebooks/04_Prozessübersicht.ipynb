```python
#######################################
# VISUALISIERUNG: Methodisches Vorgehen zur Clusteranalyse
# Darstellung als Prozess mit 5 Schritten (Top-down)
#######################################

import matplotlib.pyplot as plt
import numpy as np

# -------------------------------------
# 1. Inhalte der fünf methodischen Schritte definieren
# -------------------------------------
labels = [
    '5. Visualisierungen\n\nGeografisch, Clustereigenschaften, statistische Tests',
    '4. Modellbasierte Merkmalsanalyse\n\nRandom Forest: Wichtigste Variablen für die Clusterzugehörigkeit',
    '3. Statistische Analyse der Cluster\n\n Silhouette, Davies-Bouldin, Kruskal-Wallis, Dunn-Index:\nAnalyse der Clusterstruktur, signifikante Unterschiede zwischen Clustern,\nMessung der Trennschärfe, internen Konsistenz und Abgrenzung',
    '2. Clustering\n\nAlgorithmen: K-Means, DBSCAN, agglomeratives Clustern\nOptimierung der Hyperparameter: Clusteranzahl (K-Means), Epsilon (DBSCAN)',
    '1. Datenvorbereitung\n\nSkalierung, Kodierung, Variablenauswahl'
]

# -------------------------------------
# 2. Vorbereitung der Grafik
# -------------------------------------
fig, ax = plt.subplots(figsize=(7, 7))  # Plot-Größe

# Y-Positionen gleichmäßig verteilen (von oben nach unten)
y_pos = np.linspace(0, len(labels) - 0.5, len(labels))

# -------------------------------------
# 3. Rechteckige Textboxen zeichnen
# -------------------------------------
for i, (label, y) in enumerate(zip(labels, y_pos)):
    ax.text(
        0.5, y, label,
        ha='center', va='center', fontsize=12,
        bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3')
    )

# -------------------------------------
# 4. Formatierung der Achsen
# -------------------------------------
ax.set_xlim(-1.5, 1.5)  # Breite des Plots
ax.set_ylim(-0.5, len(labels) - 0.5)  # Höhe des Plots
ax.set_aspect('equal')  # Gleiches Seitenverhältnis (quadratisch)
# ax.set_title('Methodisches Vorgehen zur Clusteranalyse', fontsize=16)  # Optionaler Titel

# Achsenbeschriftungen und Rahmen ausblenden
ax.axis('off')

# -------------------------------------
# 5. Exportieren der Grafik
# -------------------------------------
plt.tight_layout(pad=1.5)  # Weniger Padding für engeres Layout
plt.savefig("output/plots/Prozessübersicht.png", dpi=100)
plt.show()
```

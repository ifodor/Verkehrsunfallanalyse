```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from decimal import Decimal

# -------------------------------------
# 1. Daten vorbereiten
# -------------------------------------
data = {
    "Methode": ["K-Means", "DBSCAN", "Agglomerativ"],
    "Silhouette": [0.07, 0.30, 0.18],
    "Davies-Bouldin": [2.38, 7.58, 2.05],
    "Signifikante Features": [17, 15, 149]
}
df = pd.DataFrame(data)

for container in plot1.containers:
    plot1.bar_label(container, labels=[f"{v.get_height():.2f}" for v in container], fontsize=11)

# -------------------------------------
# 2. Plot-Stil einstellen
# -------------------------------------
sns.set_style("whitegrid")

# -------------------------------------
# 3. Subplots erstellen (1 Zeile, 3 Spalten)
# -------------------------------------
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Schriftgren zentral definieren
title_size = 14
label_size = 12
tick_size = 11

# -------------------------------------
# 4. Plot 1: Silhouette-Wert
# -------------------------------------
plot1 = sns.barplot(x="Methode", y="Silhouette", hue="Methode", data=df, ax=axes[0], palette="Blues_r", legend=False)
axes[0].set_title("Silhouette-Wert (Hher ist besser)", fontsize=title_size)
axes[0].set_ylabel("Silhouette-Wert", fontsize=label_size)
axes[0].set_xlabel("")
axes[0].tick_params(axis="both", labelsize=tick_size)
for container in plot1.containers:
    plot1.bar_label(container, fmt="%.3f", fontsize=11)

# -------------------------------------
# 5. Plot 2: Davies-Bouldin-Index
# -------------------------------------
plot2 = sns.barplot(x="Methode", y="Davies-Bouldin", hue="Methode", data=df, ax=axes[1], palette="Reds_r", legend=False)
axes[1].set_title("Davies-Bouldin-Index (Niedriger ist besser)", fontsize=title_size)
axes[1].set_ylabel("Davies-Bouldin-Index", fontsize=label_size)
axes[1].set_xlabel("")
axes[1].tick_params(axis="both", labelsize=tick_size)
for container in plot2.containers:
    plot2.bar_label(container, fmt="%.2f", fontsize=11)

# -------------------------------------
# 6. Plot 3: Anzahl signifikanter Features
# -------------------------------------
plot3 = sns.barplot(x="Methode", y="Signifikante Features", hue="Methode", data=df, ax=axes[2], palette="Greens_r", legend=False)
axes[2].set_title("Anzahl signifikanter Features (Hher ist besser)", fontsize=title_size)
axes[2].set_ylabel("Anzahl Features", fontsize=label_size)
axes[2].set_xlabel("")
axes[2].tick_params(axis="both", labelsize=tick_size)
for container in plot3.containers:
    plot3.bar_label(container, fmt="%.0f", fontsize=11)

# -------------------------------------
# 7. Layout finalisieren
# -------------------------------------
plt.tight_layout()
plt.savefig("Clustering_Evaluation_Horizontal.png", dpi=100)
plt.show()
```
